# SVM Exercises

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt

from sklearn import svm
from sklearn.metrics import accuracy_score
from sklearn.svm import SVC
from sklearn import metrics
```

Load Pima diabetes dataset

```python
url = "https://raw.githubusercontent.com/BeaverWorksMedlytics2020/Data_Public/master/NotebookExampleData/Week1/diabetes.csv"
names = ['preg', 'plas', 'pres', 'skin', 'test', 'mass', 'pedi', 'age', 'class']

# YOUR CODE HERE
data = pd.read_csv(url, names=names)

# Dropping NaN rows
invalid = ['plas', 'pres', 'skin', 'test', 'mass']

for i in invalid:
    data[i].replace(to_replace=0, value=np.nan, inplace=True)
    
data = data.dropna(axis=0).reset_index(drop=True)

data.head()
```

Split data into training, colidation, and test sets with a test size of 25%

```python
# YOUR CODE HERE
X_cols = ['preg', 'plas', 'pres', 'skin', 'test', 'mass', 'pedi', 'age']

y_col = 'class'

test_size = 0.25
X_train, X_test, y_train, y_test = train_test_split(data[X_cols], data[y_col], test_size=test_size, random_state=0)

# Further split X and y of training into training and validation sets
X_train, X_val, y_train, y_val = train_test_split(X_train, y_train, test_size=test_size, random_state=0)
```

Create a model with sklearn's SVC and a high gamma value (perhaps 1.0)

```python
# YOUR CODE HERE
svm = SVC(gamma=1.0, C=1)


svm.fit(X_train, y_train)
```

Print accuracy of training and validation data

```python
# Printing accuracy of training and validation data
y_train_pred=svm.predict(X_train)
print("Training Accuracy is ", accuracy_score(y_train, y_train_pred)*100)
y_val_pred=svm.predict(X_val)
print("Validation Accuracy is ", accuracy_score(y_val,y_val_pred)*100)
```

Create a model with sklearn's SVC and a low gamma value (perhaps .001)

```python
# YOUR CODE HERE
# YOUR CODE HERE
svm2 = SVC(gamma=0.001, C=1)

# Training/fitting a model with training data
svm2.fit(X_train, y_train)
```

Print accuracy of training and validation data

```python
# YOUR CODE HERE
y_train_pred2=svm2.predict(X_train)
print("Training Accuracy is ", accuracy_score(y_train, y_train_pred2)*100)
y_val_pred2=svm2.predict(X_val)
print("Validation Accuracy is ", accuracy_score(y_val,y_val_pred2)*100)
```

Explain the accuracy results below. 

the accuracies became better because as we lowered the gamma, more datapoints were considered making the model more representative of the entire dataset

Print the confusion matrices for your high and low gamma models (both numeric and visualized)

```python
# YOUR CODE HERE
#high gamma model
print('Confusion Matrix: High')
labels = [0,1]
cm = metrics.confusion_matrix(y_val, y_val_pred)
print(cm)

#low gamma model
print('Confusion Matrix: Low')
labels = [0,1]
cmLow = metrics.confusion_matrix(y_val, y_val_pred2)
print(cmLow)
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)
cax = ax.matshow(cm,cmap='Reds')
plt.title('\n'+"High Gamma Model"+'\n', fontsize=20)
fig.colorbar(cax)
ax.set_xticklabels([''] + labels, fontsize=16)
ax.set_yticklabels([''] + labels, fontsize=16)
plt.xlabel('Predicted', fontsize=16)
plt.ylabel('True', fontsize=16)
plt.show()

fig = plt.figure()
ax = fig.add_subplot(111)
cax = ax.matshow(cmLow,cmap='Blues')
plt.title('\n'+"Low Gamma Model"+'\n', fontsize=20)
fig.colorbar(cax)
ax.set_xticklabels([''] + labels, fontsize=16)
ax.set_yticklabels([''] + labels, fontsize=16)
plt.xlabel('Predicted', fontsize=16)
plt.ylabel('True', fontsize=16)
plt.show()
```
